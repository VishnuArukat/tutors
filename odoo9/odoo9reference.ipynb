#### The steps happening when a module is installed

1. Running the preinit hook which is located in the __openerp__.py file as pre_init_hook= name of the function
        For more information odoo consist of 3 important keys which check the module has to be installed or not.which are as follows
        1.external_dependencies
        2.pre_init_hook
        3.post_init_hook

```python
#external_dependencies example is given below

'external_dependencies': {
        'python' : ['mime'],
    },
#this will check whether the module mime is installed or not,if not it will raise an error

#pre_init_hook is also used to run a function before a module is installed like we check the module compatibility like it has the same version as the module,the function is defined in the __init__.py

#in __openerp__.py

'pre_init_hook':'pre_init_check',

#in __init__.py

def pre_init_check(cr):
    from openerp.service import common
    from openerp.exceptions import Warning
    version_info = common.exp_version()
    server_serie =version_info.get('server_serie')
    if server_serie!='8.0':raise Warning('Module support Odoo series 8.0 found {}'.format(server_serie))
    return True

#post_init_hook is same as the above 

#in __openerp__.py

'post_init_hook': 'post_init_check',
 
#in __init__.py

def post_init_check(cr, registry):   
    from openerp.service import common
    from openerp.exceptions import Warning
    version_info = common.exp_version()
    server_serie =version_info.get('server_serie')
    if server_serie!='8.0':raise Warning('Module support Odoo series 8.0 found {}'.format(server_serie))
    return True
```

2. Load the model definition from the python files and update the database structure if necessary.
3. install the demo data if thedemo data is enabled.
4. Now run the **post_init_hook**
5. After that run the validation of the view definition.
6. Run the test of the addons if the tests are enabled
7. update the module state in the db
8. Update the translations in the db if any translation in the module

### Relational Fields
odoo has three major relational fields in it which are following
1. one2many
        one2many as the name refers, it is a relation to a model,just like a tree and in that we have to specify the model and the fields that we want to display in the one2many.These have no actual representation in the db.There are actually shortcuts for displaying the shortcuts.
2. many2many
    this for adding multiple records of another model to the current model. This one has additional option which can be added using the widget tag.Please refer this link https://goo.gl/fgyzLC
        this will not add any new coloumns to the db instead of that it will create a new table with the ids of the two table like they are linked to each other.The table name is created by odoo automatically by combining the two table name and adding the suffix _rel this can be overwrite by the <i>relation</i> parameter as it is more secure, if the corresponding table names are large then the resulting table name will exede the psql limit of 63 character.
    
3. many2one - this represents a many to one relation to a model which is same as dropown in html and which can have optional parameters such as ondelete,context,domain etc.
    in the db level it will store the id of the referencing model and also this will create a db constraint as we mentioned below.By default no db index is created but this can be done using the index =True.
    this type of fields have an additional parameter ,<i>auto_join</i> which can be used for one2many and many2one fields for fastening the search query on the db.For example if we have a many2one field on res.partner with domain like <br />
    <code>[('bank_ids.name', 'like', 'foo')]  </code><br/>
    this will perform 3 querys on the db as follows
1 on res_partner_bank, with domain [('name', '=', 'foo')], that returns a list of res.partner.bank ids (bids) <br/>
1 on res_partner, with a domain ['bank_ids', 'in', bids)], that returns a list of res.partner ids (pids) <br/>
1 on res_partner, with a domain [('id', 'in', pids)]<br/> 
if we use the auto_join parameter and set it to True then the odoo will join the table and make it one query to access the db this will fasten the search but will bypass the access rules.This is experimental for more information go here https://goo.gl/WTw5sq

## Heirarchy Support in odoo

Heirarchy means nested tree like structure of the database.So that there will be a parent and a child,this is called nested set structure.

## Computed Fields in Odoo
Some time you have to calculate the value of a field calculated from other fields or derived from others,for this you can use the compute fields in the odoo 9 in odoo 8 there are functional fields.Normally the these fields are set non editable but we can set it to editable.And these fields are non writabl and non searchable as default but you can use the search parameter to define a function that do the searching.
The computed fields are dynamically calculated at runtime and set a value in it,it uses caching of the values to prevent it from recalculating the value.You can use the @depends parameter to set the fields that we need to be checked for recalculating.Remember always set a default value in the computed fields to avoid the error in that field.
For enabling the write condition to a compute field we can use the inverse parameter which holds the function for write.
compute_sudo is to enable the function to run in superuser.When more than one field using the same function then you can use the multi parameter and state that they are all using the same compute function.

## Relational fields in Odoo
In the related fields these are like normal fields which have additional parameter related to that which can be linked to other model.This particularly used to fetch the data from the another model and store in it like onchange.
It can have all parameters of the normal fields and some special parameters which are in the module.To get the data for the related fields we can use the fields like many2one for accessing the data for the related fields.<br/>
Or we can use the reference fields for the comodel for the related fields as creating a function and adding the select tag for getting the comodels. <br/>
There are some referensable models that we can use for this, this enables the user to select the comodel for the related field and this can be found under the <code>Database Structure > Reference model</code> <br/>And the model for that is the 'res.request.link', in this we can add extra model reference that can be used.The crazy thing about the refernce fields that we can select the model and it will show all the records in it in a selection field next to it.It need the @api.model for to operate that is because it operates in the model level not in the record level.This may be the easy way but this type of fields will create loads if the db has lots of records for the models.Reference fields do not take advantage of the referential integrity.

## Inheritance in Odoo
inheritance of a model is the feature that can be used to modify the base models or other models without modifing the original code.it is done using the _inherit keyword or you can use the _inherits keyword for multiple models and pass the models as a list.When adding fields in the inherited models the attributes are added in the parent model and if the fields are in the parent model then the changes to that field are added not the entire field.And the methods that are declared in the parent class can be overwritten by the inherit class or can be called super and add some changes to it.
